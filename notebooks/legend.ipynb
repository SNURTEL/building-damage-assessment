```python
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
import numpy as np
```

```python
def make_legend(colors, labels, out_file, ncols=1, display=True):
    colors = [(r / 255, g / 255, b / 255) for r, g, b in colors]
    patches = [mpatches.Patch(color=c, label=l) for c, l in zip(colors, labels)]
    legend = plt.legend(handles=patches, prop={"size": 16}, framealpha=1, frameon=True, ncols=ncols)
    fig = legend.figure
    for ax in fig.axes:
        ax.axis("off")
    fig.canvas.draw()
    bbox = legend.get_window_extent()
    expand = [-4, -18, 4, 4]
    bbox = bbox.from_extents(*(bbox.extents + np.array(expand)))
    bbox = bbox.transformed(fig.dpi_scale_trans.inverted())
    fig.savefig(out_file, dpi=300, bbox_inches=bbox, transparent=True)
    print(f"Saved to {out_file}")

    if display:
        plt.show()
```

```python
xbd_colors = [(128, 128, 128), (0, 255, 0), (244, 255, 0), (255, 174, 0), (255, 0, 0)]
xbd_classes = ["Tło", "Budynki nieuszkodzone", "Niewielkie uszkodzenia", "Znaczne uszkodzenia", "Budynki zniszczone"]
make_legend(xbd_colors[1:], xbd_classes[1:], "misc/legend_xbd.png", ncols=4)
```

```python
floodnet_colors = [(128, 128, 128), (0, 255, 0), (255, 0, 0)]
floodnet_classes = ["Tło", "Budynki niezalane", "Budynki zalane"]
make_legend(floodnet_colors[1:], floodnet_classes[1:], "misc/legend_floodnet.png", ncols=2)
```

```python
floodnet_pre_colors = [
    (180, 120, 120),
    (255, 0, 0),
    (140, 140, 140),
    (160, 150, 20),
    (0, 82, 255),
    (255, 0, 245),
    (255, 235, 0),
    (4, 250, 7)
]
floodnet_pre_classes = [
    "Budynki niezalane",
    "Budynki zalane",
    "Niezalane drogi",
    "Zalane drogi",
    "Drzewa",
    "Pojazdy",
    "Baseny",
    "Trawa",
]
make_legend(floodnet_pre_colors, floodnet_pre_classes, "misc/legend_floodnet_pre.png", ncols=2)
```

```python
def hextorgb(hex):
    return tuple(int(hex.lstrip('#')[i:i+2], 16) for i in (0, 2, 4))
```

```python
example_colors = [
    hextorgb("#3C1098"),
    hextorgb("#8429F6"),
    hextorgb("#6EC1E4"),
    hextorgb("#FEDD3A"),
    hextorgb("#E2A929")
]
example_labels = [
    "Budynki",
    "Grunt",
    "Drogi",
    "Roślinność",
    "Woda"
]

make_legend(example_colors, example_labels, "misc/legend_example.png", ncols=5)
```
